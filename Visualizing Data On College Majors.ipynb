### Major Category Vs Median

```python
recent_grads.head()
```

```python
recent_grads_major_total = recent_grads[['Major', 'Total', 'Median']]
popular_majors = recent_grads_major_total.sort_values(by=['Total'], ascending=False)
high_earning_majors = recent_grads_major_total.sort_values(by=['Median'], ascending=False)
high_earning_majors.head()
```

```python
ax_total_median = popular_majors.plot(x='Total', y='Median', kind='scatter', title='Total vs Median')
```

### Findings
As seen in the scatter plot above, the students in the most popular majors tend to make less money than students in some of the least popular majors.

```python
majority_female = recent_grads[recent_grads['ShareWomen'] > 0.5]
majority_female_earnings_mean = majority_female['Median'].mean()
majority_female_earnings_median = majority_female['Median'].median()
print('Mean Earnings Majors With Majority Female: {}'.format(majority_female_earnings_mean))
print('Median Earnings Majors With Majority Female: {}'.format(majority_female_earnings_median))

majority_male = recent_grads[recent_grads['ShareWomen'] < 0.5]
majority_male_earnings_mean = majority_male['Median'].mean()
majority_male_earnings_median = majority_male['Median'].median()
print('Mean Earnings Majors With Majority Male: {}'.format(majority_male_earnings_mean))
print('Median Earnings Majors With Majority Male: {}'.format(majority_male_earnings_median))
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Median vs Share Women')
```

### Findings
The data suggests a strong negative correlation between earning money and sujects with female majorities. In other words, there are more women studying majors that make less money.
